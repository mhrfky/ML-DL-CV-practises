```python
import cv2
import numpy as np
import cv2 as cv
from matplotlib import pyplot as plt
from PIL import Image,ImageOps
from skimage import io

import urllib
```

```python
img = io.imread('https://github.com/mhrfky/ML-DL-CV-practises/blob/main/Learning/materials/coins.jpg?raw=true')
```

```python
blur = cv.blur(img,(5,5))
median = cv.medianBlur(img,5)
gaus = cv.GaussianBlur(img,(5,5),0)
bilateral = cv.bilateralFilter(img,9,75,75)
```

```python
plt.subplot(231),plt.imshow(img),plt.title('Original')
plt.xticks([]), plt.yticks([])
plt.subplot(232),plt.imshow(blur),plt.title('Blurred')
plt.xticks([]), plt.yticks([])
plt.subplot(233),plt.imshow(median),plt.title('median')
plt.xticks([]), plt.yticks([])
plt.subplot(234),plt.imshow(gaus),plt.title('Gaussian blurred')
plt.xticks([]), plt.yticks([])
plt.subplot(235),plt.imshow(bilateral),plt.title('bilateral filtering')
plt.xticks([]), plt.yticks([])

plt.show()
```

```python
img = io.imread('https://github.com/mhrfky/ML-DL-CV-practises/blob/main/Learning/materials/Noisy-image-Gaussian-noise-with-mean-and-variance-0005.png?raw=true')
img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
```

```python
blur = cv.blur(img,(5,5))
median = cv.medianBlur(img,5)
gaus = cv.GaussianBlur(img,(5,5),0)
bilateral = cv.bilateralFilter(img,9,75,75)
blur = cv2.cvtColor(blur, cv2.COLOR_GRAY2RGB)
median = cv2.cvtColor(median, cv2.COLOR_GRAY2RGB)
gaus = cv2.cvtColor(gaus, cv2.COLOR_GRAY2RGB)
bilateral = cv2.cvtColor(bilateral, cv2.COLOR_GRAY2RGB)
img = cv2.cvtColor(img, cv2.COLOR_GRAY2RGB)
```

```python
plt.subplot(231),plt.imshow(img),plt.title('Original')
plt.xticks([]), plt.yticks([])
plt.subplot(232),plt.imshow(blur),plt.title('Blurred')
plt.xticks([]), plt.yticks([])
plt.subplot(233),plt.imshow(median),plt.title('median')
plt.xticks([]), plt.yticks([])
plt.subplot(234),plt.imshow(gaus),plt.title('Gaussian blurred')
plt.xticks([]), plt.yticks([])
plt.subplot(235),plt.imshow(bilateral),plt.title('bilateral filtering')
plt.xticks([]), plt.yticks([])
plt.show()
```
